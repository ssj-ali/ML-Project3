```python
names=['a{}'.format(i) for i in range(1, 61)]
names.append('Result')
train = pd.read_csv(r'ML-Project3/train.csv', names=names)
```

```python
result=pd.get_dummies(train['Result'],drop_first=True)
train.head()
```

```python
train.drop(['Result'],axis=1,inplace=True)
train.head()
```

```python
train = pd.concat([train, result],axis=1)
```

```python
train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
names2=['a{}'.format(i) for i in range(1, 61)]
test = pd.read_csv(r'ML-Project3/test1.csv', names=names2 )
```

```python
test.head()
```

```python
combine = [train,test]
```

```python
X_train = train.drop("R",axis=1)
Y_train = train["R"]
X_train.shape , Y_train.shape
```

```python
test.info()
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, Y_train)
Y_pred = logmodel.predict(test)
acc_log = round(logmodel.score(X_train , Y_train) * 100, 2)
acc_log
```

```python
coeff_df = pd.DataFrame(train.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logmodel.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
Submission = pd.concat([test,Y_pred],axis=1)
```
